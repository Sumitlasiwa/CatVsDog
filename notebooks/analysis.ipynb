```python
import json
import matplotlib.pyplot as plt
```

```python


with open("../results/alexnet/log_data.json") as f:
    history = json.load(f)
    
# Create figure with 2 horizontal subplots
plt.figure(figsize=(16, 5))

# -------------------------------
# Loss Plot (Left)
# -------------------------------
plt.subplot(1, 2, 1)  # 1 row, 2 cols, position 1
epochs = list(range(1, len(history["train_loss"]) + 1))  # Creating list of epochs starting from 1
plt.plot(epochs, history["train_loss"], 'b.-', label="Train Loss", markersize=8)  # Blue dots
plt.plot(epochs, history["val_loss"], 'rx-', label="Val Loss", markersize=8)      # Red crosses
plt.title("Loss Curve - AlexNet", fontsize=12, pad=10)
plt.xlabel("Epochs", fontsize=10)
plt.ylabel("Loss", fontsize=10)
plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# -------------------------------
# Accuracy Plot (Right)
# -------------------------------
plt.subplot(1, 2, 2)  # 1 row, 2 cols, position 2
plt.plot(epochs, history["train_acc"], 'g.-', label="Train Acc", markersize=8)  # Green dots
plt.plot(epochs, history["val_acc"], 'mx-', label="Val Acc", markersize=8)      # Magenta crosses
plt.title("Accuracy Curve - AlexNet", fontsize=12, pad=10)
plt.xlabel("Epochs", fontsize=10)
plt.ylabel("Accuracy", fontsize=10)
plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

## Model comparison 

| Model     | Accuracy | Parameters | Epochs | Time per Epoch |
|-----------|----------|------------|--------|----------------|
| AlexNet   | 89.5%    | 61M        | 10     | 3.5 min        |
| VGG16     | 90.1%    | 138M       | 15     | 5 min          |
| ResNet18  | 92.5%    | 11M        | 7      | 3.2 min        |


We can see that the ResNet18 model has the highest accuracy among the three models, with an accuracy of 92.5%. This is likely due to its deeper architecture and the use of residual connections, which allow it to learn more complex features and improve its performance on the task. The AlexNet model has the lowest accuracy, with an accuracy of 89.5%, which is still a good result but not as good as the other two models. The VGG16 model has an accuracy of 90.1%, which is slightly better than the AlexNet model but not as good as the ResNet18 model. 
Time per epoch is also an important factor to consider, as it can affect the overall training time of the model. The ResNet18 model has the shortest time per epoch, which is 3.2 minutes, making it a good choice for training on a large dataset. The AlexNet model has a time per epoch of 3.5 minutes, which is slightly longer than the ResNet18 model. The VGG16 model has the longest time per epoch, which is 5 minutes, making it the slowest model to train. 

In conclusion, the ResNet18 model is the best choice among the three models, with the highest accuracy and the shortest time per epoch.
